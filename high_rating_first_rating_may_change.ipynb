```python
def get_min_max_rating_values() -> Dict:
    """
        returns dictionary wich contains inner dictionary with 2 keys: min and max:
        Usage:
            ratings_values = get_min_max_rating_values()
            grade_high_min_value = ratings_values['высокий']['min']
            grade_high_max_value = ratings_values['высокий']['max']
        }
    """
    min_max_values = {
        'экспресс без ФА': {
            'min': 0.02,
            'max': 0.03
        },
        'низкий': {
            'min': 0.20,
            'max': 0.22
        },
        'средний': {
            'min': 0.24,
            'max': 0.26
        },
        'высокий': {
            'min': 0.53,
            'max': 0.56
        },
    }
    return min_max_values
```

```python
def get_adjusting_coefficient(month):
    """
        Возращаем коэффициент коррекции в зависимости от полученного месяца
    """
    _date = parse(month)
    if _date < parse("01.01.2019"):
        return 0.002
    elif _date < parse("01.01.2020"):
        return 0.003
    elif _date < parse("01.01.2021"):
        return 0.004
    else:
        return 0.005
```

```python
def correct_min_max_values(rating_values, coefficient):
        rating_values['высокий']['min']          -= coefficient
        rating_values['высокий']['max']          -= coefficient
        rating_values['средний']['min']          += coefficient / 2
        rating_values['средний']['max']          += coefficient / 2
        rating_values['низкий']['min']           += coefficient / 2 - 0.001
        rating_values['низкий']['max']           += coefficient / 2 - 0.001
        # rating_values['экспресс без ФА']['min']  += coefficient / 2
        # rating_values['экспресс без ФА']['max']  += coefficient / 2
        return rating_values
```

```python
def get_rating_dependencies():
    k = 0.35
    return {
        "высокий"        : round(k * 0.04, 5),
        "средний"        : round(k * 0.08, 5),
        "низкий"         : round(k * 0.12, 5), 
        "экспресс без ФА": round(k * 0.10, 5),
    }
```

```python
def move_last_rating_to_the_end(ratings, rating_tobe_last):
    ratings[rating_tobe_last] = ratings.pop(rating_tobe_last)
    return ratings
```

```python
def get_payment_statuses(rating_dependencies):
    """
    Для корректной работы алгоритма важно, чтобы впервую очередь
    обрабатывался статус "увеличение просрочки" из категории 
    "без просрочки" в категорию "1-30".  
    """
    return {
        "выход на просрочку":{
            "Без просрочки": ("1-30"  , rating_dependencies),
        },
        "увеличение просрочки": {
            "1-30"         : ("31-60" , 0.25),
            "31-60"        : ("дефолт", 0.3),
        },
        "погашение просрочки": {
            "1-30"         : ("Без просрочки", 0.25),
            "31-60"        : ("1-30"         , 0.4),
        },
    }
```

```python
def define_rating_share(min_max_rating_values, rating, clients_with_rating, sum_month):
    global df
    global month
    _min = min_max_rating_values[rating]['min']
    sum_rating = round(df[df['ratings']==rating][month].sum(), 2)
    if sum_rating / sum_month < _min:
        without_rating = (df[month].notna()) & (df['ratings']=='без рейтинга')
        sum_rating, clients_with_rating = fill_rating(without_rating, rating, sum_rating, sum_month, _min, clients_with_rating)
    return sum_rating / sum_month
```

```python
def fill_rating(empty_criteria, rating, sum_rating, sum_month, min_value, proceed_clients):
    global df
    global month
    empty_data = df[empty_criteria]
    empty_rows_count = len(empty_data) # это быстрее, чем empty_data.shape[0] на 30 - 40%%
    cur_share = sum_rating / sum_month
    step = 1
    while cur_share < min_value and step <= empty_rows_count:
        idx = rnd.randint(0, empty_rows_count - 1)
        id_client = empty_data.index.values[idx][2]
        if id_client not in proceed_clients:
            proceed_clients[id_client] = month
        elif proceed_clients[id_client] != month:
            raise Exception ('Клиент повторно обрабатывается из предыдущего месяца!')
        else:
            continue
        df.loc[(slice(None),slice(None),id_client), 'ratings'] = rating
        sum_rating += round(df.loc[(slice(None),slice(None),id_client), month].sum(), 2)
        cur_share = sum_rating / sum_month
        step += 1
    return sum_rating, proceed_clients
```

```python
def fill_payment_overdue(rating, sum_month, target_share):
    global df
    global month
    global cur_backet_month
    cur_category = "Без просрочки" 
    target_category = "1-30"
    proceeded_clients = {}
    no_debt_df = df[(df[month].notna()) & (df[cur_backet_month]==cur_category) & (df['ratings']==rating)]
    no_debt_length = len(no_debt_df)
    try:
        sum_target_category = df.groupby([cur_backet_month, 'ratings'])[month].sum().loc[target_category, rating]
    except KeyError:
        sum_target_category = 0
    cur_share = sum_target_category / sum_month
    step = 1
    while cur_share < target_share and step <= no_debt_length:
        idx = rnd.randint(0, no_debt_length - 1)
        id_client = no_debt_df.index.values[idx][2]
        if id_client not in proceeded_clients:
            proceeded_clients[id_client] = target_category
        else:
            continue
        df.loc[(slice(None),slice(None),id_client), cur_backet_month] = target_category
        sum_target_category += round(df.loc[(slice(None), slice(None), id_client), month].sum(), 2)
        cur_share = sum_target_category / sum_month
        step += 1
    return sum_target_category
```

```python
def fill_payment_progress(cat_in_prev_backet, 
                          sum_cat_in_prev_backet, 
                          category_to_set,
                          target_share):
    global month
    global cur_backet_month
    max_count = len(cat_in_prev_backet)
    try:
        sum_cat_in_cur_backet = cat_in_prev_backet.groupby(cur_backet_month)[month].sum().loc[category_to_set]
    except KeyError:
        sum_cat_in_cur_backet = 0
    share = sum_cat_in_cur_backet / sum_cat_in_prev_backet
    step = 1
    proceeded_clients = {}
    while share < target_share and step <= max_count:
        idx = rnd.randint(0, max_count-1)
        id_client = cat_in_prev_backet.index.values[idx][2]
        if id_client not in proceeded_clients:
            proceeded_clients[id_client] = category_to_set
        else:
            continue
        df.loc[(slice(None),slice(None),id_client), cur_backet_month] = category_to_set
        sum_cat_in_cur_backet += round(df.loc[(slice(None), slice(None), id_client), month].sum(), 2)
        share = sum_cat_in_cur_backet / sum_cat_in_prev_backet
        step += 1
    return sum_cat_in_cur_backet
```

```python
def get_cur_and_prev(backets:list, month:str)->list:
    """
    Добавляем в полученный список название теекущего месяца с бакетами,
    возвращаем:
    (обновленный список, 
     название текущего столбца, 
     название предыдущего столбца)
    """
    current = 'backet_' + month
    backets.append(current)
    previous = backets[-2] if len(backets)>1 else current
    return backets, current, previous
```

```python
def fill_cur_backet_from_prev(cur_backet_month:str, prev_backet_month:str)->str:
    global df
    global month
    df[cur_backet_month] = ""
    df[cur_backet_month] = df[prev_backet_month]
    no_backet_df = df[(df[month].notna()) & (df[cur_backet_month]=="")]
    df.loc[no_backet_df.index.values, cur_backet_month] = "Без просрочки"
    res = f'Текущий {cur_backet_month} предварительно заполнен.'
    return res
```

```python
def get_rating_resulting_shares(df):
    shares = df.groupby('ratings').sum() / df.groupby('ratings').sum().sum()
    shares = pd.DataFrame(shares).T
    shares = shares.set_index(pd.to_datetime(shares.index, dayfirst=True))
    years = set([month.year for month in shares.index])
    return pd.DataFrame({year_: shares.loc[str(year_), :].mean() for year_ in years})
```

Алгоритм:
- создаем столбец для указания бакета, в который попадет платеж текущего месяца,
- для всех строк, в которых есть платеж текущего месяца, заполняем вновь созданный столбец значениями предыдущего бакета
- если в предыдущем месяце нет платежа по графику, то и бакет будет пустым, такие значения запоняем категорией "без просрочки".
- для платежей с текущей категорией "без просрочки" рандомно назначаем категорию "1-30" до тех пор, пока доля в общих платежах текущего месяца не превысит целевого значения (соотносим к сумме платежей текущего месяца);
- для иных платежей рандомно назначаем иные категории до тех пор, пока доля новой категории в сумме платежей текущего месяца, у которых предыдущая категория равна стартовой (соотносим к группировке по предыдущему бакету по категориям и сумме платежей текущего месяца);

```python
rating_values = get_min_max_rating_values()
clients_with_rating = {}
backets = []
last_rating_to_define = "экспресс без ФА"
probability_tobe_overdue = move_last_rating_to_the_end(get_rating_dependencies(), last_rating_to_define)
payment_statuses = get_payment_statuses(probability_tobe_overdue)

for idx, month in enumerate(df.columns, 1):
    try: 
        parse(month)
    except Exception:
        continue
    backets, cur_backet_month, prev_backet_month = get_cur_and_prev(backets, month)
    fill_cur_backet_from_prev(cur_backet_month, prev_backet_month)

    for status, categories in payment_statuses.items():
        if status=="выход на просрочку":
            sum_month = round(df[month].sum(), 2)
            for rating, probability in probability_tobe_overdue.items():
                # присваиваем рейтинги
                if rating!=last_rating_to_define:
                    rating_share = define_rating_share(rating_values, rating, clients_with_rating, sum_month)
                else:    
                    no_rating_df = df[(df[month].notna()) & (df['ratings']=='без рейтинга')]
                    df.loc[no_rating_df.index.values, 'ratings'] = last_rating_to_define
                # начинаем набор следующего бакета
                rating_sum = round(df[df['ratings']==rating][month].sum(), 2)
                target_share = probability
                overdue_rating_sum = fill_payment_overdue(rating, rating_sum, target_share)
                share = overdue_rating_sum / rating_sum
        elif idx > 1:
            for category, chance in categories.items():
                category_to_set = chance[0]        
                target_share = chance[1]
                cat_in_prev_backet = df[(df[month].notna()) & \
                                        (df[prev_backet_month]==category)]
                sum_cat_in_prev_backet = round(cat_in_prev_backet[month].sum(), 2)
                share_category = 0
                if sum_cat_in_prev_backet > 0:
                    sum_cat_in_cur_backet = fill_payment_progress(cat_in_prev_backet, 
                                                                  sum_cat_in_prev_backet, 
                                                                  category_to_set, 
                                                                  target_share)
                    share_category = sum_cat_in_cur_backet / sum_cat_in_prev_backet
                print(f'from {prev_backet_month} and {category}-{sum_cat_in_prev_backet:,.0f} to {cur_backet_month}-{category_to_set}-{sum_cat_in_cur_backet:,.0f}, share: {share_category:.0%}')
    k = get_adjusting_coefficient(month)                                          
    rating_values = correct_min_max_values(rating_values, k)
```
